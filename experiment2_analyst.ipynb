```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("/content/Eksperimen_HCI_ShapeItUp-Eksperimen_2.csv")

# Make sure the column names match (remove any spaces or typos)
df.columns = [c.strip().lower() for c in df.columns]

# Rename columns for consistency
df = df.rename(columns={
    'palet': 'palet',
    'jumlah kategori': 'jumlah_kategori',
    'jawaban peserta': 'jawaban_peserta',
    'jawaban benar': 'jawaban_benar',
    'status benar/salah': 'status',
    'bentuk yang dipakai': 'bentuk'
})

# Convert the status column to a numeric value: 1 (True), 0 (False)
df['accuracy'] = df['status'].map({'Correct': 1, 'Incorrect': 0})

# General statistics
print("Total jawaban:", len(df))
print("Akurasi rata-rata:", df['accuracy'].mean())

# Average accuracy per pallet
acc_palet = df.groupby('palet')['accuracy'].mean().reset_index()
plt.figure(figsize=(8, 4))
sns.barplot(data=acc_palet, x='palet', y='accuracy')
plt.title("Akurasi Rata-Rata per Palet")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.show()

# Average accuracy per number of categories
acc_kat = df.groupby('jumlah_kategori')['accuracy'].mean().reset_index()
plt.figure(figsize=(8, 4))
sns.lineplot(data=acc_kat, x='jumlah_kategori', y='accuracy', marker='o')
plt.title("Akurasi Berdasarkan Jumlah Kategori")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(True)
plt.show()

```
